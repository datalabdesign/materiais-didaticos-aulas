Este notebook foi Este notebook foi desenvolvido pelo Professor Elias Bitencourt como parte do material didático da disciplina de Métodos Digitais do curso de Design da Universidade do Estado da Bahia (UNEB).

Ele foi criado para **simplificar e automatizar** o processo de formatação de IDs de vídeos e canais para uso com o **YouTube Data Tools**. A ideia é poupar o trabalho manual de separar IDs, permitindo que o usuário simplesmente **cole a lista de IDs diretamente da tabela de resultados** (sem o cabeçalho) e obtenha o formato exigido pela interface dos módulos, como **Channel List**, **Channel Network**, **Video List**, e **Video Co-Comment Network**.

### O que este notebook faz:

1. **Entrada rápida de IDs**: O usuário pode copiar os IDs da coluna (sem cabeçalho) e colar diretamente no campo de texto.
2. **Formatação automática**: O notebook organiza e separa os IDs por vírgulas, como exigido pela interface do **YouTube Data Tools**.
3. **Remoção opcional de duplicatas**: Há uma opção para eliminar duplicatas automaticamente, caso o usuário deseje.
4. **Saída pronta**: Após o processamento, os IDs são apresentados de forma pronta para serem copiados e colados diretamente nos módulos da ferramenta.
5. **Contagem de IDs**: O notebook também conta os IDs na entrada e na saída, facilitando a verificação.

### Como usar:

1. **Cole os IDs** (copiados da coluna de resultados) no campo de texto.
2. Se preferir, marque a opção para **remover duplicatas**.
3. Clique no botão **"Formatar"** para organizar os IDs.
4. O resultado aparecerá no campo de saída, com os IDs prontos para serem usados.

```python
# @title Copiando Ids de vídeo e canais e formatando para construir redes no Youtube Data Tools
import ipywidgets as widgets
from IPython.display import display

# Criar uma área de texto para entrada
text_area = widgets.Textarea(
    value='',
    placeholder='Cole aqui os IDs dos vídeos ou canais YouTube',
    description='IDs:',
    layout=widgets.Layout(width='50%', height='200px'),
    disabled=False
)

# Checkbox para selecionar se deseja eliminar duplicatas
remove_duplicates = widgets.Checkbox(
    value=False,
    description='Remover duplicatas',
    disabled=False
)

# Área de saída com contagem
output_area = widgets.Textarea(
    value='',
    description='Resultado:',
    layout=widgets.Layout(width='50%', height='100px'),
    disabled=True
)

# Área para mostrar a contagem de IDs
count_area = widgets.Label(
    value="IDs na entrada: 0 | IDs na saída: 0"
)

# Função para processar o input e retornar os IDs separados por vírgulas
def process_input(b):
    input_text = text_area.value.strip()

    # Separar por linhas e remover espaços extras
    ids = [line.strip() for line in input_text.splitlines() if line.strip()]

    # Verificar se a checkbox de remover duplicatas está marcada
    if remove_duplicates.value:
        ids = list(set(ids))  # Remover duplicatas

    # Resultado formatado
    result = ', '.join(ids)

    # Atualizar área de saída
    output_area.value = result

    # Atualizar contagem de IDs
    count_area.value = f"IDs na entrada: {len(input_text.splitlines())} | IDs na saída: {len(ids)}"

# Botão para executar a formatação
button = widgets.Button(description="Formatar")
button.on_click(process_input)

# Mostrar os widgets: área de texto, checkbox, botão, área de saída e contagem
display(text_area, remove_duplicates, button, output_area, count_area)
```
